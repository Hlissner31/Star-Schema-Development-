```python
CreateFS = '''
   CREATE TABLE fact_sales (
    Date_Key DATE,
    SKU bigint(20),
    Store_id varchar(255),
    Quantity_Sold bigint(20),
    Revenue double,
    PRIMARY KEY (Date_Key, SKU, Store_id),
    FOREIGN KEY (Date_Key) REFERENCES dim_date(Date),
    FOREIGN KEY (SKU) REFERENCES dim_sku(SKU),
    FOREIGN KEY (Store_id) REFERENCES dim_store(store_id)
);
'''
```

```python
# read the data into a dataframe
pd.read_sql_query(CreateFS, con = buildcon)
```

```python
fs_data = '''
SELECT 
    dd.`Date`, 
    t.SKU, 
    dst.store_id, 
    SUM(t.QUANTITY) AS Quantity_Sold, 
    SUM(t.AMT) AS Revenue
FROM 
    `transactions(c)` t
JOIN 
    dim_date dd ON t.SALEDATE = dd.`Date`
JOIN 
    dim_store dst ON t.STORE = dst.store_id
WHERE 
    t.STYPE = 'P'
GROUP BY 
    dd.`Date`, 
    t.SKU, 
    dst.store_id;
    '''
```

```python
# read the data into a dataframe
df = pd.read_sql_query(fs_data, con = buildcon)
```

```python
df.to_sql(name='fact_sales', con=aws_write, if_exists='replace', index=False)
```
